### Drop highly corr col

```python
def drop_corr(df, threshold=0.9):
    corr_col = set()
    corr_matrix = df.corr()

    for i in range(len(df.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                col_name = corr_matrix.columns[i]
                corr_col.add(col_name)
    return corr_col

col_to_drop = drop_corr(df)
```

```python
# spliting the dataset

x = new_df.drop(["target"], axis=1)
y = new_df.target
```

```python
#splitting

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# scaling the data set 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_scaled_train = scaler.fit_transform(x_train)
x_scaled_test = scaler.transform(x_test)
```

# model training

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Model -1
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(x_scaled_train, y_train)
```

```python
# Make predictions on the test set
predictions = rf_clf.predict(x_scaled_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy*100: .2f}")

# Display classification report
print("Classification Report:")
print(classification_report(y_test, predictions))

print("Confution matric")
print(confusion_matrix(y_test, predictions)) 
```

```python
# Model - 2
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_scaled_train, y_train)
```

```python
# Make predictions on the test set
predictions = svc.predict(x_scaled_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy*100: .2f}")

# Display classification report
print("Classification Report:")
print(classification_report(y_test, predictions))

print("Confution matric")
print(confusion_matrix(y_test, predictions)) 
```

## hyperparameter Tunning

```python
# defining parameter range 
param_grid = {'C': [0.1, 1, 10, 100, 1000],  
            'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
            'kernel': ['rbf', 'poly']} 
```

```python
from sklearn.model_selection import GridSearchCV
svc_grid = GridSearchCV(SVC(), param_grid, verbose= 4)
```

```python
svc_grid.fit(x_scaled_train, y_train)
```

```python
# print best parameter after tuning 
print(svc_grid.best_params_) 
```

```python
# Make predictions on the test set
predictions = svc_grid.predict(x_scaled_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy*100: .2f}")

# Display classification report
print("Classification Report:")
print(classification_report(y_test, predictions))

print("Confution matric")
print(confusion_matrix(y_test, predictions)) 
```
